Question 1: What is Simple Linear Regression?

Answer 1: Simple Linear Regression is a statistical method used to model the relationship between  variable (y) and a single independent variable (x) . It assumes a linear relationship, meaning changes x correspond to proportional changes in y.

 
Question 2: What are the key assumptions of Simple Linear Regression?  
Answer 2: 1) Linearity – Relationship between x and y must be linear.
2) Independence – Observations are independent of each other.
3) Homoscedasticity – Constant variance of residuals.
4) Normality of Errors – Residuals should follow a normal distribution.
5) No Multicollinearity – Since there’s only one x multicollinearity is not relevant here.


 
Question 3: What is heteroscedasticity, and why is it important to address in regression models?
Answer 3: Heteroscedasticity :  It occurs when the variance of residuals is not constant across values of x.
-        It is problematic because it violates regression assumptions leading to:
-        Inefficient estimates of coefficients.
-        Incorrect standard errors and unreliable hypothesis tests.
-        Detecting and fixing heteroscedasticity ensures more accurate predictions and valid inference.

Question 4: What is Multiple Linear Regression?
Answer 4:  Multiple linear Regression (MLR) is an extension of simple linear Regression that models the relationship between a dependent variable (y) and two or more independent variables (x1, x2,x3….)

Question 5: What is polynomial regression, and how does it differ from linear regression?
Answer 5: Polynomial Regression is a type of regression where the relationship between the independent variable (x) and the dependent variable (y) is modeled as an nth-degree polynomial
-        Linear regression fits a straight line.
-        Polynomial regression fits a curved line making it suitable for nonlinear relationships.

Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
-  Implementation of Simple Linear Regression model

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])
model = LinearRegression()
model.fit(X,Y)
y_pred = model.predict(X)
plt.scatter(X,Y, color = 'green',label = 'Data Points')
plt.plot(X,y_pred, color = 'red',label = 'Regression Line')
```

Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000] 
● Rooms = [2, 3, 3, 4] 
● Price = [250000, 300000, 320000, 370000]


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
data = pd.DataFrame({
    'Area':  [1200, 1500, 1800, 2000] ,
 'Rooms':  [2, 3, 3, 4] ,
'Price' : [250000, 300000, 320000, 370000]
})
X = data[["Area","Rooms"]]
Y = data["Price"]
model = LinearRegression()
model.fit(X,Y)
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5] 
● Y = [2.2, 4.8, 7.5, 11.2, 14.7] 
Fit a 2nd-degree polynomial and plot the resulting curve. 


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])


# Polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Model
model = LinearRegression()
model.fit(X_poly,Y)
y_pred = model.predict(X_poly)

# Plot
plt.scatter(X,Y, color = 'green')
plt.plot(X,y_pred, color = 'red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data: 
● X = [10, 20, 30, 40, 50]
 ● Y = [15, 35, 40, 50, 65]
 Assess heteroscedasticity by examining the spread of residuals. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


X = np.array([10, 20, 30, 40, 50]).reshape(-1,1)
Y = np.array([15, 35, 40, 50, 65])


# Model
model = LinearRegression()
model.fit(X,Y)
y_pred = model.predict(X)


# Residuals
residuals = Y-y_pred


# Plot residuals
plt.scatter(X, residuals, color = 'blue')
plt.axhline(y= 0, color = 'red',linestyle = "--")
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```
